```python
df['sepal_ratio'] = df['SepalLengthCm'] / df['SepalWidthCm']
```

```python
df['petal_area'] = df.apply(lambda row:row['PetalLengthCm'] * row['PetalWidthCm'], axis = 1)
```

```python
df['SepalWidthCm'].fillna(df['SepalWidthCm'].mean(), inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Species'] = le.fit_transform(df['Species'])
```
